```python
# First let's read andBACK UP it!. Enconding corrected for UTF-8 format, No separation sep= issues.
shark = pd.read_csv('data/attacks.csv', encoding = 'latin1')
shark_backup = shark.copy()
```

```python
#CHECKING OVERALL DATASET INFORMATION TO BE AWARE OF THE SUBJECT AND MAIN POINTS, IMPORTANT TO FIRSTLY SELECT THE COLUMNS TO BE USED
#ACCORDING TO THE TYPE OF ANALISYS
shark.head(20)
```

```python
#setting column names using list comprehension
col_names = [col.lower().replace(' ','_').replace('.','_').replace(':','_') for col in shark.columns]
shark.columns = col_names
shark.columns
```

```python
# dropping useless columns for the specific analisys, remaining from 24 only 14 columns.
shark.drop(['case_number','name','investigator_or_source','pdf','href_formula','href','case_number_1','case_number_2','unnamed__22','unnamed__23'], axis=1, inplace=True)
shark
```

```python
# Ordering columns and changing on column names as desired 
shark.columns
col_order = ['original_order','year','date','time', 'type', 'country', 'area', 'location', 'activity',
     'sex_', 'age', 'injury', 'fatal_(y/n)' , 'species_']
len(col_order)
shark = shark.loc[:,col_order]
shark.rename(columns={'original_order':'case_number'},inplace=True)
shark
```

```python
#Deleting all unnecessary full null rows, and inplacing it.
shark.dropna(how='all',inplace=True)
```

```python
# Verify the previous full null rows dropping procedure
shark.info()
```

```python
#cheking if there are full duplicated rows
shark[shark.duplicated()]
```

```python
# Listing the aquatic attack activity  types in order of occurrence, counting values in the column
activity_count = shark['activity'].value_counts()
activity_count
```

```python
#creating a mask to list the 5 most often aquatic activities attack cases using lamba funtion
shark['activity'].apply(lambda x: x if x in activity_count.head().index else np.nan)
```

```python
#creating a column showing the only the 5 most often aquatic activities cases, using lambda function
shark['major_activity'] = shark['activity'].apply(lambda x: x if x in activity_count.head().index else np.nan)
shark['major_activity'] 
```

```python
# Listing the 30 most often injury types in order of occurrence, counting values in the column
injury_count = shark['injury'].value_counts()
injury_count.head(60)
```

```python
##creating a mask to list the 5 most often aquatic activities cases using lambda function
shark['injury'].apply(lambda x: x if x in injury_count.head(30).index else np.nan)
```

```python
#creating a column showing only the 5 most often aquatic activities cases, using lamba function
shark['major_injury'] = shark['injury'].apply(lambda x: x if x in injury_count.head(30).index else np.nan)
shark['major_injury']
```

```python
#finding the fatal cases proportion per attack activity type using groupby function. (major_activity / major_injury )calculation
shark_count = shark.groupby(by=['major_activity']).count().reset_index()
shark_count 
```

```python
# Listing the countries in order of occurrence counting values in the column, to answer question n.4
country_count = shark['country'].value_counts() 
country_count.head(5)
```
